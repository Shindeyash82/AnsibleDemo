```python
import tensorflow as tf
from tqdm import tqdm # This library adds a nice progress bar

# Define our image dimensions. We already know this from our test.
IMG_WIDTH = 128
IMG_HEIGHT = 128
IMG_CHANNELS = 3 # Satellite images are RGB

# --- Define the function to load all data ---

def load_data(img_dir, mask_dir, filenames):
    # Create empty numpy arrays to hold the data
    # (num_files, height, width, channels)
    X = np.zeros((len(filenames), IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS), dtype=np.float32)

    # (num_files, height, width, 1 channel)
    y = np.zeros((len(filenames), IMG_HEIGHT, IMG_WIDTH, 1), dtype=np.float32)

    print(f"Loading data from {img_dir} and {mask_dir}...")

    # Use tqdm for a progress bar
    for n, filename in enumerate(tqdm(filenames)):

        # --- Load Image ---
        img_path = os.path.join(img_dir, filename)
        image = cv2.imread(img_path, cv2.IMREAD_COLOR)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        # Resize (just in case) and normalize
        image = cv2.resize(image, (IMG_WIDTH, IMG_HEIGHT))
        X[n] = image / 255.0  # Normalize to 0-1

        # --- Load Mask ---
        mask_name = filename.replace("image_", "mask_")
        mask_path = os.path.join(mask_dir, mask_name)
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        # Resize, normalize, and add the channel dimension
        mask = cv2.resize(mask, (IMG_HEIGHT, IMG_WIDTH))
        mask = mask / 255.0 # Normalize to 0-1
        # Convert to 0s or 1s (in case some pixels are gray)
        mask = (mask > 0.5).astype(np.float32)
        y[n] = np.expand_dims(mask, axis=-1)

    return X, y

print("Helper function defined.")
```

```python
# --- Load Training Data ---
TRAIN_IMG_DIR = "dataset/train/images/"
TRAIN_MASK_DIR = "dataset/train/masks/"
# Get the list of all .png filenames
train_filenames = sorted([f for f in os.listdir(TRAIN_IMG_DIR) if f.endswith('.png')])
# Load the data
X_train, y_train = load_data(TRAIN_IMG_DIR, TRAIN_MASK_DIR, train_filenames)

print(f"\nTraining data loaded!")
print(f"X_train (images) shape: {X_train.shape}")
print(f"y_train (masks) shape: {y_train.shape}")

# --- Load Validation Data ---
VAL_IMG_DIR = "dataset/validation/images/"
VAL_MASK_DIR = "dataset/validation/masks/"
# Get the list of all .png filenames
val_filenames = sorted([f for f in os.listdir(VAL_IMG_DIR) if f.endswith('.png')])
# Load the data
X_val, y_val = load_data(VAL_IMG_DIR, VAL_MASK_DIR, val_filenames)

print(f"\nValidation data loaded!")
print(f"X_val (images) shape: {X_val.shape}")
print(f"y_val (masks) shape: {y_val.shape}")

print("\n--- Step 3 Complete: All data is loaded and ready for the model! ---")
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Conv2DTranspose, concatenate, Dropout

def build_unet_model(input_shape):
    """
    Builds a U-Net model for segmentation.
    """
    inputs = Input(input_shape)

    # --- Encoder (Contracting Path) ---
    # Block 1
    c1 = Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(inputs)
    c1 = Dropout(0.1)(c1)
    c1 = Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c1)
    p1 = MaxPooling2D((2, 2))(c1)

    # Block 2
    c2 = Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p1)
    c2 = Dropout(0.1)(c2)
    c2 = Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c2)
    p2 = MaxPooling2D((2, 2))(c2)

    # --- Bottleneck ---
    c_mid = Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(p2)
    c_mid = Dropout(0.2)(c_mid)
    c_mid = Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c_mid)

    # --- Decoder (Expansive Path) ---
    # Block 3
    u3 = Conv2DTranspose(32, (2, 2), strides=(2, 2), padding='same')(c_mid)
    u3 = concatenate([u3, c2]) # This is the "skip connection"
    c3 = Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u3)
    c3 = Dropout(0.1)(c3)
    c3 = Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c3)

    # Block 4
    u4 = Conv2DTranspose(16, (2, 2), strides=(2, 2), padding='same')(c3)
    u4 = concatenate([u4, c1]) # This is the "skip connection"
    c4 = Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(u4)
    c4 = Dropout(0.1)(c4)
    c4 = Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(c4)

    # --- Output Layer ---
    # This final layer maps all the features to a single 1-channel image
    # We use 'sigmoid' activation for binary classification (pixel is 0 or 1)
    outputs = Conv2D(1, (1, 1), activation='sigmoid')(c4)

    # Create the model
    model = Model(inputs=[inputs], outputs=[outputs])
    return model

print("U-Net model function defined.")
```

```python
# Define the input shape using our constants
input_shape = (IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS)

# Build the model
model = build_unet_model(input_shape)

# Compile the model
# We use 'binary_crossentropy' because each pixel is either 0 (no landslide) or 1 (landslide)
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Print the model summary
print("Model built and compiled successfully!")
print(model.summary())
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# Define where to save the best model
best_model_path = "landslide_segmentation_model.keras"

# Define the callbacks
callbacks = [
    # Stop training if the validation loss doesn't improve for 3 epochs
    EarlyStopping(patience=3, monitor='val_loss', verbose=1),

    # Save the best model (lowest val_loss) to the file
    ModelCheckpoint(filepath=best_model_path,
                    monitor='val_loss',
                    save_best_only=True,
                    verbose=1)
]

print("Callbacks defined!")
```

```python
# --- Start Training! ---

# We will run for 20 epochs, but EarlyStopping will likely stop it sooner.
EPOCHS = 20
BATCH_SIZE = 16

# This command starts the training process
history = model.fit(
    X_train,
    y_train,
    validation_data=(X_val, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks
)

print("\n--- Training Complete! ---")
print(f"The best model has been saved to {best_model_path}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pickle

# --- 1. Define Features (X) and Target (y) ---
target_column = 'Landslide'

# All columns EXCEPT the target
feature_columns = [col for col in df.columns if col != target_column]

# Columns that need scaling (the float64 ones)
continuous_features = [
    'Rainfall_mm', 'Slope_Angle', 'Soil_Saturation',
    'Vegetation_Cover', 'Earthquake_Activity', 'Proximity_to_Water'
]

X = df[feature_columns]
y = df[target_column]

# --- 2. Split Data ---
# We use 'stratify=y' to make sure both train and val sets
# have a similar percentage of 0s and 1s.
X_train, X_val, y_train, y_val = train_test_split(
    X, y,
    test_size=0.2,        # 20% for validation, 80% for training
    random_state=42,      # Ensures we get the same split every time
    stratify=y
)

# --- 3. Scale Data ---
scaler = StandardScaler()

# IMPORTANT: Fit the scaler ONLY on the training data
X_train[continuous_features] = scaler.fit_transform(X_train[continuous_features])

# Use that same scaler to transform the validation data
X_val[continuous_features] = scaler.transform(X_val[continuous_features])

print("Data successfully split and scaled!")
print(f"Training data shape: {X_train.shape}")
print(f"Validation data shape: {X_val.shape}")

# --- 4. Save the Scaler ---
# This is CRITICAL. We need this file for our API.
SCALER_SAVE_PATH = 'oracle_data_scaler.pkl'
with open(SCALER_SAVE_PATH, 'wb') as f:
    pickle.dump(scaler, f)

print(f"\nScaler saved to {SCALER_SAVE_PATH}. You will need to download this file.")
```

```python
# We will use the Sequential API, which is simpler and will fix the error.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

def build_dnn_model(input_shape):
    """
    Builds a Deep Neural Network (DNN) for tabular data
    using the Sequential API.
    """
    model = Sequential([
        # This layer defines the input shape (9 features)
        InputLayer(input_shape=input_shape, name='feature_input'),

        # Hidden layers
        Dense(64, activation='relu'),
        Dropout(0.2),
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(64, activation='relu'),

        # Output layer: 1 neuron, sigmoid activation (for 0 or 1)
        Dense(1, activation='sigmoid', name='output')
    ])

    # Compile the model
    # We use 'binary_crossentropy' because our target is 0 or 1.
    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    return model

print("DNN model function (Sequential version) defined.")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Rescaling, Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.layers import RandomFlip, RandomRotation, RandomZoom

# --- Define the Data Augmentation Layers ---
# This "pre-processor" will run on the GPU and create new
# versions of your images for every epoch.
data_augmentation = Sequential([
    InputLayer(shape=(IMG_SIZE, IMG_SIZE, 3)), # Input is 224x224x3
    RandomFlip("horizontal"),
    RandomRotation(0.1), # Rotate up to 10%
    RandomZoom(0.1),     # Zoom in up to 10%
], name="data_augmentation")


# --- Define the Model ---
model = Sequential([
    # --- 1. Augmentation & Preprocessing ---
    # Add the augmentation layers first
    data_augmentation,

    # Normalize pixel values from [0, 255] to [0, 1]
    Rescaling(1./255, name="rescaling"),

    # --- 2. The CNN "Body" (Feature Extractor) ---
    # Block 1
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2)),

    # Block 2
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2)),

    # Block 3
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2)),

    # --- 3. The "Head" (Classifier) ---
    # Flatten the 3D feature maps into a 1D vector
    Flatten(),

    # A "hidden" layer with 128 neurons
    Dense(128, activation='relu'),
    Dropout(0.5), # Dropout prevents overfitting

    # The final output layer
    # 1 neuron = binary classification (damage vs. no_damage)
    # 'sigmoid' activation = outputs a probability (0.0 to 1.0)
    Dense(1, activation='sigmoid')
], name="damage_classifier_model")

print("Model architecture defined.")
```

```python
# Compile the model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy', # Perfect for 2-class (0/1) problems
    metrics=['accuracy']      # Show us the accuracy
)

# Print the full summary
print("Model built and compiled successfully!")
model.summary()
```
